```python
import pandas as pd

def obtener_datos_acciones_tesla():
    tesla = yf.Ticker("TSLA")
    return tesla.history(period="max")

datos_acciones_tesla = obtener_datos_acciones_tesla()
datos_acciones_tesla.head()
```

```python
import requests
from bs4 import BeautifulSoup

print("\nQuestion 2 - Extracting Tesla Revenue Data Using Webscraping:")

def obtener_datos_ingresos_tesla():
    url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
    payload = {}
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36'}
    response = requests.request("GET", url, headers=headers, data=payload)
    # response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    tabla = soup.find('table', {'class': 'table'})
    filas = tabla.find_all('tr')
    datos_ingresos = []
    for fila in filas[1:]:
        columnas = fila.find_all('td')
        año = columnas[0].text
        ingreso = columnas[1].text.replace('$', '').replace(',', '')
        datos_ingresos.append((año, ingreso))
    return datos_ingresos

# Ejecución
datos_ingresos_tesla = obtener_datos_ingresos_tesla()
df_ingresos_tesla = pd.DataFrame(datos_ingresos_tesla, columns=["Año", "Ingresos"])
df_ingresos_tesla.tail()
```
